```python
import pandas as pd
import numpy as np
import datetime
df1=pd.read_excel("/content/202107_Sale_All.xlsx")
```

1. Chọn các cột thông tin cần thiết phục vụ báo cáo phân tích

```python
select_column = ['Mã hàng','Tên hàng','Ngày','Số phiếu','Lý do','Số lượng','Đơn giá','Đơn giá sau thuế','Thuế suất','Chiết khấu($)','Thành tiền','Thành tiền trước thuế cộng chiết khẩu','Nhân viên','Mã NV','Kho','Khách hàng','Mã KH','Chi nhánh','Thuế','Hình thức TT','Hạng thẻ','CK KHTT(%)','CK(%)','Doanh thu','Nhóm C1','Nhóm C2','Nhóm C3']
df2=df1[select_column]
```

2. Kiểm tra định dạng dữ liệu, đổi lại định dạng đúng

```python
#df2.head()
df2.info()
df2['Kho']=df2['Kho'].astype('object')
```

3. Thay các giá trị null cột CK KHTT(%) và cột Hạng thẻ

```python
df2['CK KHTT(%)'].fillna(value=0,inplace=True)
df2['Hạng thẻ'].fillna('Khách hàng mới',inplace=True)
df2.info()
```

4. Tách cột chi nhánh

```python
df2[['Store','Store 1']]=df2['Chi nhánh'].str.split('#',expand=True)
df2.drop(columns=['Store 1','Chi nhánh'],inplace=True)
df2
```

5.   Split cột Ngày






```python
df2['Ngày bán hàng'] = pd.to_datetime(df2['Ngày']).dt.date
df2['Giờ bán hàng'] = pd.to_datetime(df2['Ngày']).dt.time.apply(lambda x: x.replace(microsecond=0))
df2.drop(columns=['Ngày'],inplace = True)
```

```python
df2['Ngày bán hàng'] = df2['Ngày bán hàng'].astype('datetime64[ns]')
df2.info()
```



6.   Kho = 8.04 thì thay dữ liệu ở cột store từ ONL thành ONL HCM




```python
#df2.head()
df2.loc[df2['Kho'] == 8.04, 'Store'] = 'ONL HCM'
df2[df2['Kho'] == 8.04].head()
```
